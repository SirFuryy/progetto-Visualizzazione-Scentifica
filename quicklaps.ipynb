Una volta ottenuti tutti i gp che ci interessano, otteniamo tutte le poleposition e le uniamo in un unico dataset

```python
plt.rc('figure', figsize=(15.0, 10.0))
mpl.rcParams['figure.facecolor'] = 'None' # Il colore di sfondo del grafico seguirà quello del PowerPoint
mpl.rcParams['axes.facecolor'] = 'None'
f1plt.setup_mpl(color_scheme=None)

colors_palette = {
    'Red Bull': '#CCA9DD',
    'Ferrari': '#CB3234',
    'Mercedes': '#A3E7D6',
    'Alpine': '#99D6EA',
    'Kick Sauber': '#641C34',
    'McLaren': '#EFA94A',
    'Haas': '#987654',
    'RB': '#1E213D',
    'Williams': '#4592CE',
    'Aston Martin': '#216477',
    '2024': '#CDA434',
    '2023': '#ABCDEF',
    'SQ': '#177245'
}

# Create empty lists to store the fastest lap times and GP numbers
fastest_laps_2023 = []
fastest_laps_2024 = []

# List of sessions for 2023 and 2024
sessions_2023 = [G1_2023, G2_2023, G3_2023, G4_2023, G5_2023, G6_2023, G7_2023, G8_2023, G9_2023, G11_2023, G12_2023, G13_2023, G14_2023, G16_2023]
sessions_2024 = [G1_2024, G2_2024, G3_2024, G4_2024, G5_2024, G6_2024, G7_2024, G8_2024, G9_2024, G11_2024, G12_2024, G13_2024, G14_2024, G16_2024]
circuit_name = ['01-BAR', '02-SAU', '03-AUS', '04-MIA', '05-MON', '06-ESP', '07-AUT', '08-GBR', '09-HUN', '11-ITA', '12-SGP', '13-USA', '14-MEX', '16-VEG']
```

```python
# Extract fastest lap times for 2023
for i, session in enumerate(sessions_2023, start=1):
    fastest_lap = session.laps.pick_fastest()
    fastest_laps_2023.append({'GP': circuit_name[i-1], 'Fastest Lap Time': fastest_lap['LapTime']})

# Extract fastest lap times for 2024
for i, session in enumerate(sessions_2024, start=1):
    fastest_lap = session.laps.pick_fastest()
    fastest_laps_2024.append({'GP': circuit_name[i-1], 'Fastest Lap Time': fastest_lap['LapTime']})

# Convert lists to DataFrames
df_fastest_laps_2023 = pd.DataFrame(fastest_laps_2023)
df_fastest_laps_2024 = pd.DataFrame(fastest_laps_2024)

plt.figure(figsize=(12, 6))

# Plot fastest lap times for 2023
plt.plot(df_fastest_laps_2023['GP'], df_fastest_laps_2023['Fastest Lap Time'].apply(lambda x: x.total_seconds()), 'o', color= colors_palette['2023'], label='2023')

# Plot fastest lap times for 2024
plt.plot(df_fastest_laps_2024['GP'], df_fastest_laps_2024['Fastest Lap Time'].apply(lambda x: x.total_seconds()), 'o', color= colors_palette['2024'], label='2024')

# Add labels and title
plt.xticks(rotation=45)
plt.xlabel('GP Number')
plt.ylabel('Fastest Lap Time')
plt.title('Fastest Lap Times for 2023 and 2024 Seasons')
plt.legend()

# Show plot
plt.show()
```

```python

differ = df_fastest_laps_2023['Fastest Lap Time'] - df_fastest_laps_2024['Fastest Lap Time']
difference = differ.apply(lambda x: x.total_seconds() * 10)  # Convert to decimi di secondo
# Plot the difference
plt.figure(figsize=(12, 6))
plt.plot(df_fastest_laps_2024['GP'], difference, 'o', color=colors_palette['2024'], label='2024')

# Add labels and title
plt.xticks(rotation=45)
plt.xlabel('Gran Premio (numero - Nome)')
plt.ylabel('Differenza in Decimi di Secondo')
plt.title('Differenza del giro veloce in ogni gran premio con pari condizioni tra 2023 e 2024')
plt.axhline(0, color=colors_palette['2023'], linestyle='--')  # Add a horizontal line at y=0
plt.ylim(-5, 30)
plt.legend() 

# Show plot
plt.show()
```

```python
# Plot the difference as a divergent bar chart
plt.figure(figsize=(12, 6))
bars = plt.bar(df_fastest_laps_2024['GP'], difference, color=[colors_palette['2024'] if val > 0 else colors_palette['2023'] for val in difference])

# Add labels and title
plt.xticks(rotation=45)
plt.xlabel('Gran Premio (numero - Nome)')
plt.ylabel('Differenza in Decimi di Secondo')
plt.title('Differenza del giro veloce in ogni gran premio con pari condizioni tra 2023 e 2024')
plt.axhline(0, color='black', linestyle='--')  # Add a horizontal line at y=0
plt.ylim(-3, 3)

# Show plot
plt.show()
# Plot the difference as a horizontal divergent bar chart
plt.figure(figsize=(12, 6))
bars = plt.barh(df_fastest_laps_2024['GP'], difference, color=[colors_palette['2024'] if val > 0 else colors_palette['2023'] for val in difference])

# Add labels and title
plt.yticks(rotation=0)
plt.ylabel('Gran Premio (numero - Nome)')
plt.xlabel('Differenza in Decimi di Secondo')
plt.title('Differenza del giro veloce in ogni gran premio con pari condizioni tra 2023 e 2024')
plt.axvline(0, color='black', linestyle='--')  # Add a vertical line at x=0
plt.xlim(-3, 3)

# Show plot
plt.show()
# Plot the difference as an overlapping area chart
plt.figure(figsize=(12, 6))
plt.fill_between(df_fastest_laps_2024['GP'], difference, where=(difference > 0), interpolate=True, color=colors_palette['2024'], alpha=0.5, label='2024 Faster')
plt.fill_between(df_fastest_laps_2024['GP'], difference, where=(difference <= 0), interpolate=True, color=colors_palette['2023'], alpha=0.5, label='2023 Faster')

# Add labels and title
plt.xticks(rotation=45)
plt.xlabel('Gran Premio (numero - Nome)')
plt.ylabel('Differenza in Decimi di Secondo')
plt.title('Differenza del giro veloce in ogni gran premio con pari condizioni tra 2023 e 2024')
plt.axhline(0, color='black', linestyle='--')  # Add a horizontal line at y=0
plt.ylim(-3, 3)
plt.legend()

# Show plot
plt.show()
```
