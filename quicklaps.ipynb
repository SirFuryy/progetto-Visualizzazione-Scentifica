Una volta ottenuti tutti i gp che ci interessano, otteniamo tutte le poleposition e le uniamo in un unico dataset

```python
# Create empty lists to store the fastest lap times and GP numbers
fastest_laps_2023 = []
fastest_laps_2024 = []

# List of sessions for 2023 and 2024
sessions_2023 = [G1_2023, G2_2023, G3_2023, G4_2023, G5_2023, G6_2023, G7_2023, G8_2023, G9_2023, G10_2023, G11_2023, G12_2023, G13_2023, G14_2023, G16_2023]
sessions_2024 = [G1_2024, G2_2024, G3_2024, G4_2024, G5_2024, G6_2024, G7_2024, G8_2024, G9_2024, G10_2024, G11_2024, G12_2024, G13_2024, G14_2024, G16_2024]
circuit_name = ['01-BAR', '02-SAU', '03-AUS', '04-MIA', '05-MON', '06-ESP', '07-AUT', '08-GBR', '09-HUN', '10-BEL', '11-ITA', '12-SGP', '13-USA', '14-MEX', '16-VEG']

# Extract fastest lap times for 2023
for i, session in enumerate(sessions_2023, start=1):
    fastest_lap = session.laps.pick_fastest()
    fastest_laps_2023.append({'GP': circuit_name[i-1], 'Fastest Lap Time': fastest_lap['LapTime']})

# Extract fastest lap times for 2024
for i, session in enumerate(sessions_2024, start=1):
    fastest_lap = session.laps.pick_fastest()
    fastest_laps_2024.append({'GP': circuit_name[i-1], 'Fastest Lap Time': fastest_lap['LapTime']})

# Convert lists to DataFrames
df_fastest_laps_2023 = pd.DataFrame(fastest_laps_2023)
df_fastest_laps_2024 = pd.DataFrame(fastest_laps_2024)

print(df_fastest_laps_2023)
print(df_fastest_laps_2024)

plt.figure(figsize=(12, 6))

# Plot fastest lap times for 2023
plt.plot(df_fastest_laps_2023['GP'], df_fastest_laps_2023['Fastest Lap Time'].apply(lambda x: x.total_seconds()), marker='o', label='2023')

# Plot fastest lap times for 2024
plt.plot(df_fastest_laps_2024['GP'], df_fastest_laps_2024['Fastest Lap Time'].apply(lambda x: x.total_seconds()), marker='x', label='2024')

# Add labels and title
plt.xticks(rotation=45)
plt.xlabel('GP Number')
plt.ylabel('Fastest Lap Time')
plt.title('Fastest Lap Times for 2023 and 2024 Seasons')
plt.legend()

# Show plot
plt.show()
```

```python
# Calculate the difference in decimi di secondo between 2024 and 2023 fastest lap times
df_fastest_laps_2024['Difference (decimi)'] = -(df_fastest_laps_2024['Fastest Lap Time'] - df_fastest_laps_2023['Fastest Lap Time']) * 10

# Plot the difference
plt.figure(figsize=(12, 6))
plt.plot(df_fastest_laps_2024['GP'], df_fastest_laps_2024['Difference (decimi)'], marker='o', linestyle='-', color='b')

# Add labels and title
plt.xticks(rotation=45)
plt.xlabel('GP Number')
plt.ylabel('Difference in Decimi di Secondo')
plt.title('Difference in Fastest Lap Times between 2024 and 2023 Seasons')
plt.axhline(0, color='r', linestyle='--')  # Add a horizontal line at y=0

# Show plot
plt.show()
```
