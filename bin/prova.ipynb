```python
import matplotlib.pyplot as plt
import numpy as np
import matplotlib as mpl
import fastf1.plotting as f1plt

plt.rc('figure', figsize=(15.0, 10.0))
mpl.rcParams['figure.facecolor'] = 'None' # Il colore di sfondo del grafico seguirà quello del PowerPoint
mpl.rcParams['axes.facecolor'] = 'None'
f1plt.setup_mpl(color_scheme=None)

# Dati di esempio
auto = ['Auto A', 'Auto B', 'Auto C']
velocita_anno1 = [200, 220, 220]
velocita_anno2 = [210, 230, 210]

x = np.arange(len(auto))  # Posizioni delle barre
larghezza = 0.4  # Larghezza delle barre

# Creazione delle barre
plt.figure(figsize=(12, 6))
plt.figure(figsize=(10, 6))
plt.bar(x - larghezza/2, velocita_anno1, larghezza, label='Anno 1', color='blue')
plt.bar(x + larghezza/2, velocita_anno2, larghezza, label='Anno 2', color='orange')

# Etichette e titolo
plt.xticks(x, auto)
plt.xlabel('Automobili')
plt.ylabel('Velocità Massima (km/h)')
plt.title('Confronto Velocità Massima tra Due Anni')
plt.legend()
plt.show()
```

```python
# Dati di esempio
plt.figure(figsize=(12, 6))
plt.plot(auto, velocita_anno1, marker='o', label='Anno 1', color='blue')
plt.plot(auto, velocita_anno2, marker='o', label='Anno 2', color='orange')

# Etichette e titolo
plt.xlabel('Automobili')
plt.ylabel('Velocità Massima (km/h)')
plt.title('Confronto Velocità Massima')
plt.legend()
plt.show()
```

```python
# Calcolo della differenza
differenze = np.array(velocita_anno2) - np.array(velocita_anno1)

# Grafico
plt.figure(figsize=(12, 6))
plt.bar(auto, differenze, color=['green' if d > 0 else 'red' for d in differenze])
plt.axhline(0, color='black', linewidth=0.8, linestyle='--')
plt.xlabel('Automobili')
plt.ylabel('Differenza di Velocità (km/h)')
plt.title('Differenze di Velocità Massima tra Due Anni')
plt.show()
```

```python
# Calcolo della differenza
differenze = np.array(temperature_questanno) - np.array(temperature_annoscorso)

# Grafico delle differenze
plt.figure(figsize=(12, 6))
plt.bar(giorni, differenze, color=['red' if d >= 0 else 'blue' for d in differenze])
plt.axhline(0, color='black', linewidth=0.8, linestyle='--')
plt.xlabel('Giorni')
plt.ylabel('Differenza di Temperatura (°C)')
plt.ylim(-5, 5)
plt.title('Differenze di Temperatura: Quest\'anno vs Anno scorso')
plt.show()
```

```python
import seaborn as sns
import pandas as pd

# Creazione di un DataFrame
df = pd.DataFrame({
    'Giorni': giorni,
    'Quest_anno': temperature_questanno,
    'Anno_scorso': temperature_annoscorso
})

# Heatmap
plt.figure(figsize=(12, 6))
sns.heatmap(df.set_index('Giorni'), cmap='coolwarm', annot=True, fmt=".1f")
plt.title('Confronto delle Temperature (Heatmap)')
plt.show()
```

```python
# Grafico ad area sovrapposto
plt.figure(figsize=(12, 6))
plt.fill_between(giorni, temperature_questanno, alpha=0.5, label="Quest'anno", color='red')
plt.fill_between(giorni, temperature_annoscorso, alpha=0.5, label="Anno scorso", color='blue')

# Dettagli del grafico
plt.xlabel('Giorni')
plt.ylabel('Temperatura (°C)')
plt.title('Confronto delle Temperature: Aree Sovrapposte')
plt.ylim(35, 45)
plt.legend()
plt.show()
```
